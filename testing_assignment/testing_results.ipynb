## 1. Testing Your Code

### Automated vs. Manual Testing
* **Manual Testing**: Exploring the app by hand. Checking features without a fixed plan.
* **Automated Testing**: Using scripts to run a test plan. Much faster and more consistent than human testing.

### Unit Tests vs. Integration Tests
* **Unit Test**: Checks a small, single component (like a function) to isolate bugs.
* **Integration Test**: Checks how multiple parts work together (e.g., car battery, alternator, and lights).

```python
# testing sum() with a list
def test_sum():
    assert sum([1, 2, 3]) == 6, "Should be 6"

if __name__ == "__main__":
    test_sum()
    print("Everything passed")
```

Testing Multiple Data Types
Testing sum() with a tuple instead of a list. I'm using incorrect values here to trigger an AssertionError and see how the script handles a mistake.

```python
# testing sum() with tuple (expected failure)
def test_sum_tuple():
    assert sum((1, 2, 2)) == 6, "Should be 6"

if __name__ == "__main__":
    test_sum_tuple()
    print("Everything passed")
```

## 2. Writing Your First Test

### The Test Workflow
* **Create inputs**: Defining the initial data to be tested.
* **Execute code**: Running the function and saving the output.
* **Compare**: Matching the result against the expected answer.

### Common `unittest` Assertions
| Method | Equivalent to |
| :--- | :--- |
| `.assertEqual(a, b)` | `a == b` |
| `.assertTrue(x)` | `bool(x) is True` |
| `.assertIn(a, b)` | `a in b` |
| `.assertIsInstance(a, b)` | `isinstance(a, b)` |

```python
import unittest
from my_sum import sum

class TestSum(unittest.TestCase):
    def test_list_int(self):
        """
        Test that it can sum list of integers
        """
        data = [1, 2, 3]
        result = sum(data)
        self.assertEqual(result, 6)

if __name__ == '__main__':
    # Using argv=[''] to prevent errors inside a Notebook
    unittest.main(argv=[''], verbosity=2, exit=False)
```

## 3. Executing Your First Test

### Test Runners
The **test runner** is the application that executes the test code, checks assertions, and provides results. Python's built-in runner is `unittest`.

### Test Discovery and CLI
In a standard environment, tests can be run via the command line:
* `python -m unittest test`: Runs a specific module named `test`.
* `python -m unittest -v test`: Runs with **verbose** output, showing specific test names.
* `python -m unittest discover`: Automatically finds and runs all files matching `test*.py`.

### Understanding Output
* **.** (Dot): Represents a passing test.
* **F** (Fail): Represents a test where an assertion failed.
* **E** (Error): Represents a test where the code itself crashed.

```python
import unittest
from my_sum import sum
from fractions import Fraction

class TestSum(unittest.TestCase):
    def test_list_int(self):
        """
        Test that it can sum a list of integers
        """
        data = [1, 2, 3]
        result = sum(data)
        self.assertEqual(result, 6)

    def test_list_fraction(self):
        """
        Test that it can sum a list of fractions
        """
        data = [Fraction(1, 4), Fraction(1, 4), Fraction(2, 5)]
        result = sum(data)
        self.assertEqual(result, 1)

if __name__ == '__main__':
    unittest.main(argv=[''], verbosity=2, exit=False)
```

### Interpreting the Failure
The output shows **FAIL: test_list_fraction**. 

* **Traceback**: Shows exactly which line failed (`self.assertEqual(result, 1)`).
* **AssertionError**: Explains the mismatch: `Fraction(9, 10) != 1`.
* **Summary**: Indicates `Ran 2 tests` with `FAILED (failures=1)`.

This tells me that while the integer sum works, the fraction sum failed since my expected result (1) didn't match the actual math (0.9).

### Summary and Interpretation of Test Results

I ran two unit tests using the `unittest` runner to check my `sum()` function. The results show 1 pass and 1 failure:

* **test_list_int**: This test passed, shown by the **dot (.)** in the output. It confirms the function correctly handles a basic list of integers.
* **test_list_fraction**: This test failed, shown by the **F**.

**What these results mean:**
The failure in the second test prompted an `AssertionError: Fraction(9, 10) != 1`. This means my func. calculated the math correctly (1/4 + 1/4 + 2/5 = 0.9), but my test was written to expect the result to be 1. These results demonstrate that the test runner is working exactly as it should. It caught a discrepancy between the output and the expected result. This confirms that automated unit testing is a great way to isolate specific parts of my code to make certain that each component does exactly as intended.
